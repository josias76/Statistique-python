## Traitement des valeurs manquantes

NaN est une valmeur définie pour répresenteer quelque chose qui n'est pas un nombre (Not a Number) alors que son type l'exige.


```python
# Détermination des colonnes avec des valeurs manquante
titanic_df.isnull().sum()
```

```python
# remplacement des valeurs manquantes dans la colonnes 'age'
titanic_df_age = titanic_df.fillna(value = {'age': 0})
```

```python
titanic_df_age.isnull().sum()
```

```python
# rremplacement par la moyenne d'age
titanic_df_age2 = titanic_df.fillna(value = {'age': titanic_df["age"].mean()})
```

```python
titanic_df_age2.isna().sum()
```

```python
# supression des lignes manquantes 
titanic_sans_na = titanic_df.dropna()
```

### Agrégation

```python
# Exemple d'agrégation
titanic_df.groupby('pclass')["age"].mean()
```

```python
#pour avoir 2 ou 3 fonction/agrégé suivant plusieurs fonction

titanic_df.groupby('pclass')["age"].agg(["mean","median"])
```

```python
# agreger suivant plusieurs colonnes
titanic_df.groupby(['pclass','sexe'])["age"].agg(["mean","median"])
```
